* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_df = "Resources/schools_complete.csv"
student_data_df = "Resources/students_complete.csv"

school_data = pd.read_csv(school_data_df)
school_data.head()

student_data = pd.read_csv(student_data_df)
student_data.head()


```

```python
school_data_complete= pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.to_csv('school_data_complete', index=False)
school_df = school_data_complete
school_df
```

```python
total_school=len(pd.value_counts(school_df['school_name']))

total_school

```

```python
total_student=sum(pd.value_counts(school_df['student_name']))
total_student
```

```python
#Calculate the total budget
total_budget = sum(school_data['budget'])
total_budget
```

```python
#Calculate the average math score
average_math =round(school_df['math_score'].mean(),6)
average_math
```

```python
#Calculate the average reading score
average_reading =round(school_df['reading_score'].mean(),6)
average_reading
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
stud_passing_math = school_df.loc[school_df["math_score"] >= 70]
num_stud_passing_math = stud_passing_math["Student ID"].count()

pct_passing_math = (num_stud_passing_math / total_student) * 100

pct_passing_math
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)
stud_passing_reading = school_df.loc[school_df["reading_score"] >= 70]
num_stud_passing_reading = stud_passing_reading["Student ID"].count()

pct_passing_reading = (num_stud_passing_reading / total_student) * 100

pct_passing_reading
```

```python
#Calculate the percentage of students who passed math and reading (% Overall Passing)
Overall_passing = school_df[(school_df['math_score'] >= 70) & (school_df['reading_score'] >= 70)]['Student ID'].count()/total_student*100
Overall_passing
```

```python
#Create a dataframe to hold the above results
district_summary = pd.DataFrame({
    "Total Schools": total_school,
    "Total Students": f"{total_student:,}",
    "Total Budget": f"${total_budget:,.2f}",
    "Average Math Score": f"{average_math:.6f}",
    "Average Reading Score": f"{average_reading:.5f}",
    "% Passing Math": f"{pct_passing_math:.6f}",
    "% Passing Reading": f"{pct_passing_reading:.6f}",
    "% Overall Passing": f"{Overall_passing: .6f}"
}, index=[0])

district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#School Name, total students, total school budget and average math score and reading score

school_groupby = school_df.groupby("school_name")
school_mean = school_groupby.mean()

school_budget= school_mean['budget']
school_total_students = school_mean['size']
school_reading = school_mean['reading_score']
school_math = school_mean['math_score']

#student type
school_type = school_data.set_index('school_name')
school_type = school_type['type']
school_type

#per student budget
per_student_budget = school_budget / school_total_students

#percent passing Math 

school_summary = school_df[school_df['math_score']>=70].groupby(['school_name'])
percent_passing_math = (school_summary.math_score.count() / school_total_students)*100

#percent passing reading
school_summary = school_df[school_df['reading_score']>=70].groupby(['school_name'])
percent_passing_reading = (school_summary.reading_score.count() / school_total_students)*100


#calculate the overall percentage passing
#to get the overall percentage we need % for both math and reading 
#group.count by school name and student from main df

percent_math_reading = school_df[(school_df["reading_score"] >= 70)
                                               & (school_df["math_score"] >= 70)]

percent_math_reading = percent_math_reading.groupby(["school_name"]).count()["student_name"]
overall_percent = percent_math_reading / school_total_students * 100


school_summary = pd.DataFrame({
                                "School Type" : school_type,
                                "Total Students" : school_total_students,
                                "Total School Budget" : school_budget,
                                "Per Student Budget": per_student_budget,
                                "Average Math Score" : school_math,
                                "Average Reading Score" : school_reading,
                                "% Passing Math" : percent_passing_math,
                                "% Passing Reading" : percent_passing_reading,
                               "% Overall Passing" : overall_percent
                            })


school_summary
```

```python
#Top Performing Schools (By % Overall Passing)
#Sort and display the top five performing schools by % overall passing.

top_performing_schools = school_summary.sort_values(["% Overall Passing"], ascending=False)
top_performing_schools.head()
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
  

```python
# create the table to hold spending ranges
# four spending bins and group by school names.

bins = [0, 585, 615, 645, 675]

group_names = ["<$584", "$585-630", "$630-645", "$645-680"]

school_summary['Spending Ranges (Per Student)'] = pd.cut(per_student_budget, bins, labels=group_names)

# Calculate spending averages for all land put in them in the desired columns. 
#average spending ranges for math_score
spending_math = school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]

#average spending ranges for reading_score
spending_reading = school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]

#average spending ranges for percent passing_math
spending_passing_math = school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]

#average spending ranges percent passing_reading
spending_passing_reading = school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]

#average spending ranges for the overall passing percentage
overall_spending = school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]

spending_summary_df = pd.DataFrame({
          "Average Math Score" : spending_math,
          "Average Reading Score": spending_reading,
          "% Passing Math": spending_passing_math,
          "% Passing Reading": spending_passing_reading,
          "% Overall Passing": overall_spending})

spending_summary_df
```

## Scores by School Size



* Perform the same operations as above, based on school size.


```python
# create the table to hold school sizes

size_bins = [0, 1000, 2000, 5000]

group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]


#using the column total students to determine the school size 
school_summary["School Size"] = pd.cut(school_summary["Total Students"], size_bins, labels=group_names)

size_math = school_summary.groupby(["School Size"]).mean()["Average Math Score"]

size_reading = school_summary.groupby(["School Size"]).mean()["Average Reading Score"]

size_passing_math = school_summary.groupby(["School Size"]).mean()["% Passing Math"]

size_passing_reading = school_summary.groupby(["School Size"]).mean()["% Passing Reading"]

overall_passing_size = school_summary.groupby(["School Size"]).mean()["% Overall Passing"]


size_summary_df = pd.DataFrame({
                                "Average Math Score" : size_math,
                                "Average Reading Score": size_reading,
                                "% Passing Math": size_passing_math,
                                "% Passing Reading": size_passing_reading,
                                "% Overall Passing": overall_passing_size
                                })

size_summary_df


```
